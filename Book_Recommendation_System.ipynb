### Popularity based recommendation System

```python
ratings_with_books = ratings.merge(books,on = 'ISBN')
```

In this dataframe, we have different ISBN numbers for the same book title also.

So, we can't work with ISBN number..

```python
ratings_with_books.groupby('Book-Title').count()
```

```python
num_of_ratings_on_book = ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_of_ratings_on_book.rename(columns={'Book-Rating':"num_of_ratings"},inplace=True)
num_of_ratings_on_book
```

```python
avg_ratings_on_book = ratings_with_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings_on_book.rename(columns={'Book-Rating':"avg_ratings"},inplace=True)
avg_ratings_on_book
```

```python
popularity_df = num_of_ratings_on_book.merge(avg_ratings_on_book,on='Book-Title')
popularity_df
```

I have to find the books with more than 250 ratings..

As i will use it as a parameter for popularity of book

```python
popular_df = popularity_df[popularity_df['num_of_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

For showing it on the website i need the everything about the book on the books dataset.



```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates("Book-Title")
```

```python
popular_df.to_csv("Top_popular_books.csv")
```

### Collaborative filtering based recommender system

finding the name of the users who have given ratings to more than 200 books.

```python
x = ratings_with_books.groupby("User-ID").count()["Book-Rating"] > 100
users_with_threshold_ratings = x[x].index
```

```python
user_rating_filter = ratings_with_books[ratings_with_books['User-ID'].isin(users_with_threshold_ratings)]
```

Finding the books with >= 50 ratings

```python
y = user_rating_filter.groupby("Book-Title").count()['Book-Rating'] >= 50
book_with_threshold_ratings = y[y].index
book_rating_filter = user_rating_filter[user_rating_filter['Book-Title'].isin(book_with_threshold_ratings)]
book_rating_filter
```

```python
pt = book_rating_filter.pivot_table(index = "Book-Title", columns='User-ID', values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

so we have 1070 unique books and 1634 unique users according to our threshold values

finding the similarity of one book wrt to others

```python
from sklearn.metrics.pairwise import cosine_similarity
sim_score = cosine_similarity(pt)
sim_score
```

Making a recommend function

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(sim_score[index])), key = lambda x: x[1], reverse = True)[1:11]
    # return similar_items
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Zoya')
```
